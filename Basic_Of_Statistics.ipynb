#**Q9)Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).**
#ANS:-
#The normal distribution is a symmetric, bell-shaped curve characterized by two parameters: the mean (center) and the standard deviation (spread). Its key properties are:

# Symmetry around the mean  the left and right halves are mirror images.
# Mean = median = mode  all located at the center.
# Total area under the curve = 1, representing the whole probability.
# Tails approach but never touch the horizontal axis (they extend infinitely).

#The empirical rule (68-95-99.7 rule) tells us how data is distributed in a normal distribution:
#* 68% of the data falls within 1 standard deviation of the mean.
#* 95% falls within 2 standard deviations.
#* 99.7% falls within 3 standard deviations.
